```python
import geopandas as gpd
import pandas as pd

df_poll = pd.read_csv("data/pollution/export_moyenne.csv", sep=";")

gdf_poll = gpd.GeoDataFrame(
    df_poll,
    geometry=gpd.points_from_xy(df_poll["Longitude"], df_poll["Latitude"]),
    crs="EPSG:4326"
)

gdf_poll = gdf_poll[gdf_poll["Polluant"] == "PM2.5"]  # on garde PM2.5

```

```python
df_rev = pd.read_csv("data/revenus/FILO2019_DISP_COM.csv", sep=";")

df_rev_clean = df_rev[["CODGEO", "NBMEN19", "Q219"]].rename(
    columns={"Q219": "revenu_median"}
)

# On garde uniquement les communes (codes à 5 chiffres)
df_communes = df_rev_clean[df_rev_clean["CODGEO"].str.len() == 5]

```

```python
gdf_com = gpd.read_file("data/géographie/communes/a_com2021_2154.shp")
gdf_com.head()
```

```python
gdf_join = gpd.sjoin(
    gdf_poll_2154,
    gdf_com[["codgeo", "libgeo", "geometry"]],
    how="left",
    predicate="within"   # le point est dans le polygone de la commune
)

gdf_join.head()
gdf_join[["nom site", "Polluant", "valeur", "codgeo", "libgeo"]].head()

```

```python
df_pm25_com = (
    gdf_join
    .groupby("codgeo")["valeur"]
    .mean()
    .reset_index()
    .rename(columns={"valeur": "pm25_moyen"})
)

df_pm25_com.head()
```

```python
df_communes = df_rev_clean[df_rev_clean["CODGEO"].str.len() == 5]
```

```python
df_communes = df_communes.copy()
df_communes.loc[:, "CODGEO"] = df_communes["CODGEO"].astype(str)

df_pm25_com["codgeo"] = df_pm25_com["codgeo"].astype(str)
```

```python
df_final = gdf_join.merge(
    df_communes[["CODGEO", "revenu_median"]],
    left_on="codgeo",
    right_on="CODGEO",
    how="left"
)
```
